## Numpy Array Indexing and Slicing

```python
import numpy as np
import csv

with open('AAPL-2017.csv',newline='') as f:
    reader = csv.DictReader(f)
    data2017 = np.empty((0,6))
    withDate2017 = np.empty((0,7))
    #print(np.shape(data2017))
    #print(np.ndim(data2017))
    for row in reader:
        oneRow = np.array([[row['Open'], row['High'], row['Low'], row['Close'], row['Adj Close'], row['Volume']]])
        data2017 = np.append(data2017,oneRow,axis=0)
        row2 = np.array([[row['Date'],row['Open'], row['High'], row['Low'], row['Close'], row['Adj Close'], row['Volume']]])
        withDate2017 = np.append(withDate2017,row2,axis=0)
        
print(withDate2017)
```

```python
with open('AAPL-2016.csv',newline='') as f:
    reader = csv.DictReader(f)
    data2016 = np.empty((0,6))
    withDate2016 = np.empty((0,7))
    #print(np.shape(data2016))
    #print(np.ndim(data2016))
    for row in reader:
        oneRow = np.array([[row['Open'], row['High'], row['Low'], row['Close'], row['Adj Close'], row['Volume']]])
        data2016 = np.append(data2016,oneRow,axis=0)
        row2 = np.array([[row['Date'],row['Open'], row['High'], row['Low'], row['Close'], row['Adj Close'], row['Volume']]])
        withDate2016 = np.append(withDate2016,row2,axis=0)
        
print(withDate2016)
    
```
